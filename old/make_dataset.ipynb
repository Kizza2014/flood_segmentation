```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("lihuayang111265/flood-semantic-segmentation-dataset")
print("Path to dataset files:", path)
path = os.path.join(path, 'dataset')
```

Convert image from png to jpg:

```python
from PIL import Image
import os

train_path = os.path.join(path, 'train')
val_path = os.path.join(path, 'val')
img_save_path = 'data/images'
mask_save_path = 'data/masks'

train_img = os.listdir(os.path.join(train_path, 'images'))
for filename in train_img:
    img = Image.open(os.path.join(train_path, 'images', filename))
    img = img.convert('RGB')
    new_name = os.path.splitext(filename)[0] + '.jpg'
    img.save(os.path.join(img_save_path, new_name), 'JPEG')
    print(f'Converted image {filename}')
```

```python
val_img = os.listdir(os.path.join(val_path, 'images'))
for filename in val_img:
    img = Image.open(os.path.join(val_path, 'images', filename))
    img = img.convert('RGB')
    new_name = os.path.splitext(filename)[0] + '.jpg'
    img.save(os.path.join(img_save_path, new_name), 'JPEG')
    print(f'Converted image {filename}')
```
